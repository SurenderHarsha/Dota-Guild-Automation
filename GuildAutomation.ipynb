```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
import pygetwindow as gw
import time
import win32api
import win32con
import pywinauto
import pyautogui as pa
import pandas as pd
import numpy as np
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
def focus_to_window(window_title=None):
    window = gw.getWindowsWithTitle(window_title)[0]
    if window.isActive == False:
        pywinauto.application.Application().connect(handle=window._hWnd).top_window().set_focus()
```

```python
df = pd.DataFrame(columns = ['Name','GP','Delta','CumulativeD'])
```

```python
df
```

```python
pa.PAUSE = 0.1
focus_to_window('Dota 2')
checked_names = []
for k in range(8):
    #focus_to_window('Dota 2')
    pa.moveTo(5,5,duration=1)
    im1 = pa.screenshot()
    
    cvim = cv2.cvtColor(np.array(im1), cv2.COLOR_RGB2BGR)
    
    
    crop_img = cvim[370:370+542, 280:280+630]

    grayImage = cv2.cvtColor(crop_img, cv2.COLOR_BGR2GRAY)
  
    (thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
    crop_img = blackAndWhiteImage
    
    rotated = crop_img[:,90:]
    #cv2.imwrite('im_'+str(k)+'.png',rotated)
    hist = cv2.reduce(rotated,1, cv2.REDUCE_AVG).reshape(-1)

    th = 2
    H,W = rotated.shape[:2]
    uppers = [y for y in range(H-1) if hist[y]<=th and hist[y+1]>th]
    lowers = [y for y in range(H-1) if hist[y]>th and hist[y+1]<=th]
    
    for j in range(len(uppers)):
        try:
            if lowers[j]-uppers[j]<8:
                print("Skipped")
                continue
            i = rotated[uppers[j]:lowers[j],:]
        except Exception as e:
            print(e)
            continue
        name = i[:,:203]
        g = i[:,413:]
        
        small = cv2.resize(name, (0,0), fx=15, fy=15,interpolation=cv2.INTER_CUBIC)
        txt = pytesseract.image_to_string(small,config='--psm 10',lang='eng')
        large = cv2.resize(g, (0,0), fx=15, fy=15,interpolation=cv2.INTER_CUBIC)
        txt2 = pytesseract.image_to_string(large,config='--psm 10',lang='eng')
        gp = [str(s) for s in txt2 if s.isdigit()]
        gp = ''.join(gp)
        try:
            gp = int(gp)
        except Exception as e:
            print(e)
            continue
        if txt in checked_names:
            continue
        checked_names.append(txt)
        
        if txt not in df['Name'].tolist():
            #if gp in df['GP'].tolist():
            #    continue
            print("Added",txt,"To database.")
            df.loc[df.shape[0]] = [txt,gp,0,0]
            continue
        
        data = df.loc[df['Name'] == txt]
        df.loc[df['Name'] == txt ,'Delta'] = int(gp-data['GP'])
        df.loc[df['Name'] == txt, 'GP'] = gp
        df.loc[df['Name'] == txt, 'CumulativeD'] += int(gp-data['GP'])
        
    
    pa.moveTo(600,600,duration=1)

    pa.scroll(-300,x=600,y=600)
    #scroll(-3,delay_between_ticks=0.5)
    
    
```
